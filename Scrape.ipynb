```python

import requests
from bs4 import BeautifulSoup
 
 
# Making a GET request
r = requests.get('https://path.upmc.edu/cases/dxindex.html') 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

content = []

for s in soup.find_all('tr'):
    i = s.find_all('a')
    try:
        print(i[0].text)
        content.append(i[0].text)
    except:
        pass
#     for x in i:
#         content.append(x.text)
```

```python
import pandas as pd
df = pd.DataFrame(content)
```
